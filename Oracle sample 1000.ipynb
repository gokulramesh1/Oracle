```python
from datetime import date
```

```python
def weekend(dt):
    ymd = dt.split('T')[0].split('-')
    y, m, d = ymd[0], ymd[1].lstrip('0'), ymd[2].lstrip('0')
    y, m, d = int(y), int(m), int(d)
    if date(y, m, d).weekday() > 3:
        return 1
    else:
        return 0
```

```python
def get_c(c):
    return c[8:].split("'")[0]
```

```python
df['certifications'] = df['certifications'].apply(get_c)
```

```python
def convert_cert(c):
    if c == 'U/A':
        return 1
    elif c == 'U':
        return 2
    elif c == 'A':
        return 3
    else:
        return 0
```

```python
df['certifications'] = df['certifications'].apply(convert_cert)
```

```python
def convert_time(s):
    a = 0
    if s[-2] == 0:
        a = s[-1]
    else:
        a = s[-2:]
    return int(s[:2].lstrip('0')) + round(int(a)/60, 1)
```

```python
df['show_time'] = df['show_time'].apply(convert_time)
```

```python
def impute_capacity(cols):
    cap = cols[0]
    facility = cols[1]
    if pd.isnull(cap):
        if facility == 'Multiplex':
            return 260
        elif facility == 'Single Screen':
            return 530
        elif facility == 'Complex':
            return 290
    else:
        return cap
```

```python
df['capacity'] = df[['capacity', 'facility_type']].apply(impute_capacity, axis=1)
```

```python
def convert_facility(f):
    if f == 'Multiplex':
        return 1
    elif f == 'Single Screen':
        return 2
    else:
        return 3
```

```python
df['facility_type'] = df['facility_type'].apply(convert_facility)
```
