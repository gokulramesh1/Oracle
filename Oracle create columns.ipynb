```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Dataframes\\Oracle\\data_frame.csv")
```

```python
df.head()
```

```python
df = df[:3000]
```

```python
columns = ['capacity', 'cast', 'certifications', 'city', 'crew', 'experiences', 'facility_type', 'language', 'province', 'runningTime', 'sold_seats', 'time', 'date', 'releaseDates']
```

```python
a = ast.literal_eval(df['movie_data'][0])
```

```python
for key in a:
    if key in columns:
        df[key] = None

for i in range(len(df)):
    d = ast.literal_eval(df['movie_data'][i])
    for key in d:
        if key in columns:
            df[key][i] = d[key]
```

```python
df.head()
```

```python
df['cast'] = df['cast'].apply(lambda x: [actor['name'] for actor in x])    
```

```python
def director(crew):
    try:
        return crew[0]['roles'][0]['name']
    except:
        return

df['director'] = df['crew'].apply(director)
```

```python
b = ast.literal_eval(df['session_data'][0])
```

```python
for key in b['session']:
    if key in columns:
        df[key] = None

for i in range(len(df)):
    d = ast.literal_eval(df['session_data'][i])
    d = d['session']
    for key in d:
        if key in columns:
            df[key][i] = d[key]
```

```python
df['date'] = df['session_data'].apply(lambda x: ast.literal_eval(x)['session']['date'])
```

```python
df.head()
```

```python
df = df.loc[:, 'releaseDates':'experiences']
```

```python
df.drop('crew', axis=1, inplace=True)
```

```python
df.head()
```

```python
def get_date(release):
    try:
        return release['IN']
    except:
        return
```

```python
df['release_date'] = df['releaseDates'].apply(get_date)
```
